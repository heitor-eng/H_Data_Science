##EDA - Exploratory Data Analysis

EDA - Exploratory Data Analysis

To start the exploratory analysis of the dataset, let's look at the graph of two variables, the  `new_cases_smoothed` and the `new_deaths_smoothed`. These two variables can show the growth or decay of new cases and new deaths.

The reason for that is because these two variables are calculated through the formula of the **sample moving average** of these cases during seven days, as shown in the procedure below:

$$SMA = \frac{d_1 + d_2 + d_3 + ... + d_n}{n}$$

**Where:**

> **d** = The total of new cases or new deaths of the period *n*

> **n** = The number of total periods, for this case is 7

```python
# Configuring the style
sns.set_style('ticks')

# build the figure
fig, (ax1,ax2) =plt.subplots(nrows= 2,ncols = 1, figsize=(14,14))

# variable n_cases_smooth
n_cases_smooth = (df_w.loc[df_w.location =="World", ['date','new_cases_smoothed']].reset_index(drop=True))

# Graph about the number of cases smoothed
sns.lineplot(x='date',y = 'new_cases_smoothed', data=n_cases_smooth, ci = None, linewidth=5, color = 'Grey', ax = ax1)
ax1.set_title('\n Moving Average Smoothed of the number of New Cases of COVID-19 in the World \n', color = 'Grey', size = 20,loc = 'left')
ax1.set_xlabel([],color = 'white')
ax1.set_ylabel([],color = 'white')
ax1.yaxis.set_ticklabels(['0','0','200k','400k','800k','1M'])

# variable about the number of new deaths smoothed
n_deaths_smooth =(df_w.loc[df_w.location =='World', ['date','new_deaths_smoothed']].reset_index(drop=True))

# Graph about the new deaths
sns.lineplot(x='date',y = 'new_deaths_smoothed', data=n_deaths_smooth, ci = None, linewidth=5, color = 'DarkBlue', ax = ax2)
ax2.set_title('\n Moving Average Smoothed of the number of New Deaths of COVID-19 in the World \n', color = 'Grey', size = 20,loc = 'left')
ax2.set_xlabel([],color = 'white')
ax2.set_ylabel([],color = 'white')
ax2.yaxis.set_ticklabels(['0','0','2k','4k','6k','8k','10k','12k','14k'])

sns.despine()
plt.tight_layout()
```

**At first sight**

The interpretation of these two graphs are:
* The number of the variable `new_cases` is increasing again
* The difference between the number of `new_cases` and `new_deaths` is quite different
* The number of the variable `new_deaths` seems to be starting to maintain a pattern, without many peaks and valleys

These three points are what we can see through these graphs.

**Reproduction Rate (R0)** :

It is the rate that shows how fast is the spread of the vírus at the local. 
> "If the population average R0 is greater than 1, the infection will spread exponentially. If the R0 is less than 1, the infection will only spread slowly, and eventually disappear. By [Oxford Brasil](https://oxfordbrazilebm.com/index.php/quando-isso-acabara-uma-introducao-aos-numeros-de-reproducao-viral-r0-e-re/).

### Waffle Plot - A different way to show your data

According to the web page [Our World in Data](https://ourworldindata.org/covid-vaccinations?country=OWID_WRL) at the date Dec 8th, 2021, we have that approximately 55% of the world population has been vaccinated, either partial or complete, while 45% have not. As is show  at the image below:

<p align=center><img src="https://github.com/heitor-eng/H_Data_Science/blob/master/Imagens/coronavirus-data-explorer.png?raw=true" width="80%"></p>

We can pass the same information using the **Waffle Plot**, but here I need to use the proportion of cases, like:
> 3.55 billion are equivalent to 44% of the data, while 811.9 million are equal to 11% and with that, we have 55% that take vaccines and the other 45% are the 4.36 billion that are not vaccines.

To build the graph, I will use a sample of 100 cases here, respecting the same proportion as above for cases of people who were fully vaccinated, partially, and without having been vaccinated. 

```python
# install pywaffle
!pip install pywaffle -q
```

```python
# Calling the libraries
import matplotlib.pyplot as plt
from pywaffle import Waffle
import numpy as np
```

```python
fig = plt.figure(
    FigureClass=Waffle, 
    rows=5, 
    figsize = (16,4),
    values={'Fully vaccinated': 44, 'Only partly vaccinated': 11, 'Did not take any dose': 45}, 
         title={
        'label': 'Number of people vaccinated against COVID-19\n',
        'loc': 'left',
        'fontdict': {
            'fontsize': 20
        }
    },
    colors=("OliveDrab", "DarkBlue","DarkRed"),
    icons = ['syringe','syringe','skull'],
    legend={'loc': 'upper left', 
            'bbox_to_anchor': (1, 1),
            'fontsize':20
            }, 
    icon_size=24, 
    icon_legend=True
)

fig.set_tight_layout(False)
plt.show();
```
